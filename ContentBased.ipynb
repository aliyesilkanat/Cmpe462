```python
import pandas as pd
from scipy.sparse import csr_matrix
import numpy as np
```

```python
userRatings=pd.read_csv("ml-latest-small/ratings.csv",usecols=["userId","movieId","rating"])
userRatings.head()
```

```python
movies=pd.read_csv("ml-latest-small/movies.csv",usecols=["movieId","title"])
movies.head()
```

```python
ratings = pd.merge(movies, userRatings)

ratings.head()
```

```python

ratingMatrix = ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
ratingMatrix.head()
```

```python
#min_period to throw out results where fewer than 100 users rated a given movie pair
corrMatrix = ratingMatrix.corr(method='pearson', min_periods=100) 
corrMatrix.head()
```

```python
oneUserRatings = ratingMatrix.loc[50].dropna() #ratings of user, having id 50
oneUserRatings
```

```python
simCandidates = pd.Series()
for i in range(0, len(oneUserRatings)):
    print ("Adding sims for " + oneUserRatings.index[i] + "...")
    # Retrieve similar movies to this one that user rated
    sims = corrMatrix[oneUserRatings.index[i]].dropna()
    # Now scale its similarity by how well movie is rated
    sims = sims.map(lambda x: x * oneUserRatings[i])
    # Add the score to the list of similarity candidates
    simCandidates = simCandidates.append(sims)
    
#Glance at our results so far:
simCandidates = simCandidates.groupby(simCandidates.index).sum() #some movies are similiar to more than one movie, so take sum of the similiarity
simCandidates.sort_values(inplace = True, ascending = False)

simCandidates
```

```python
#Drop previously rated movies
filteredSims = simCandidates[~simCandidates.index.isin(oneUserRatings.index)]
filteredSims
```
